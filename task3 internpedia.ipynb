```python


import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Unemployment in India.csv')

# Remove extra spaces from column names
data.columns = data.columns.str.strip()

# Check the column names
print(data.columns)

# Convert the 'Date' column to datetime format
if 'Date' in data.columns:
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)
else:
    print("The column 'Date' is not present in the dataframe.")

# Plot the unemployment rate over time for each region
for region in data['Region'].unique():
    region_data = data[data['Region'] == region]
    plt.plot(region_data.index, region_data['Estimated Unemployment Rate (%)'])
    plt.title(f'Unemployment Rate in {region} over Time')
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.show()

# Calculate the average unemployment rate for each region
average_unemployment_rates = data.groupby('Region')['Estimated Unemployment Rate (%)'].mean()
print(average_unemployment_rates)

# Calculate the maximum and minimum unemployment rates for each region
max_unemployment_rates = data.groupby('Region')['Estimated Unemployment Rate (%)'].max()
min_unemployment_rates = data.groupby('Region')['Estimated Unemployment Rate (%)'].min()
print(max_unemployment_rates)
print(min_unemployment_rates)

# Plot the average unemployment rate for each region
plt.bar(average_unemployment_rates.index, average_unemployment_rates.values)
plt.xlabel('Region')
plt.ylabel('Average Unemployment Rate (%)')
plt.title('Average Unemployment Rate by Region')
plt.show()
```
